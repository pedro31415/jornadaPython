```python
import pandas as pd

table = pd.read_csv("clientes.csv")

display(table)
```

```python
display(table.info())
```

```python
# LabelEnconder

from sklearn.preprocessing import LabelEncoder

# Exemplo

# cientista -> 1
# bombeiro -> 2
# engenheiro -> 3

# profissao

condificador_profissao = LabelEncoder()
table["profissao"] = condificador_profissao.fit_transform(table["profissao"])

# mix_credito

condificador_credito = LabelEncoder()
table["mix_credito"] = condificador_credito.fit_transform(table["mix_credito"])
# comportamento_pagamento

condificador_pagamento = LabelEncoder()
table["comportamento_pagamento"] = condificador_pagamento.fit_transform(table["comportamento_pagamento"])

display(table.info())

display(table)
```

```python
# y -> coluna de previsão 
y = table["score_credito"]

# x -> as colunas da base de dados
x = table.drop(columns=["score_credito", "id_cliente"])
```

```python
# Passo 3 -> Treinar a Inteligencia Artificial

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)

# Arvore de decisao
# Nearest Neighbros -> vizinhos proximos

# importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# criar a IA

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# treinar a IA

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Escolher o melhor modelo
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# acuracia
from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# Usar o melhor modelo para fazer previsao de novos clientes

tabela_novos_clientes = pd.read_csv("novos_clientes.csv")

tabela_novos_clientes["profissao"] = condificador_profissao.transform(tabela_novos_clientes["profissao"])

# mix_credito

tabela_novos_clientes["mix_credito"] = condificador_credito.transform(tabela_novos_clientes["mix_credito"])

# comportamento_pagamento

tabela_novos_clientes["comportamento_pagamento"] = condificador_pagamento.transform(tabela_novos_clientes["comportamento_pagamento"])

nova_previsao = modelo_arvoredecisao.predict(tabela_novos_clientes)
display(nova_previsao)
```
